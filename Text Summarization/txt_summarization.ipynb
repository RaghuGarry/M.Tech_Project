```python
import spacy
from spacy.lang.en.stop_words import STOP_WORDS
from string import punctuation
```

```python
#We use Extractive Summarization in this Project
```

```python
#The Stop_words avaiable are...
stopwords = list(STOP_WORDS)
stopwords
```

```python
with open("F:/Project-M.Tech/pgm/input.txt", "r") as file:
    text = file.read()
```

```python
print(text)
```

```python
nlp=spacy.load('en_core_web_sm')
doc=nlp(text)
```

```python
tokens=[token.text for token in doc]
print(tokens)
#Note that punctuations and stop_words are also the part of tokens
#We remove them from the token 
```

```python
#The punctuations available are...
punctuation
```

```python
#If we want a new punctuation to be added .. we can add that by..
#punctuation = punctuation + '\n'
punctuation = punctuation + '\n'
```

```python
#Word Frequency Table
word_freq={}
for word in doc:
    if word.text.lower() not in stopwords:
        if word.text.lower() not in punctuation:
            if word.text not in word_freq.keys():
                word_freq[word.text]=1
            else:
                word_freq[word.text]+=1
```

```python
word_freq
```

```python
max_freq=max(word_freq.values())
max_freq
```

```python
#Normalising the freq
for word in word_freq.keys():
    word_freq[word]=word_freq[word]/max_freq
```

```python
#Normalized word_freq's....
word_freq
```

```python
#Senetence Tokenization
sentence_tokens=[sent for sent in doc.sents]
sentence_tokens
```

```python
print(len(sentence_tokens))#Total no.of sentenses present
```

```python
#Calculating sentence_scores
sentense_scores={}
for sent in sentence_tokens:
    for word in sent:
        if word.text.lower() in word_freq.keys():
            if sent not in sentense_scores.keys():
                sentense_scores[sent]=word_freq[word.text.lower()]
            else:
                sentense_scores[sent]+=word_freq[word.text.lower()]
```

```python
#Sentense Score for each of the sentense
sentense_scores
```

```python
from heapq import nlargest
```

```python
#We want to extract the 40% percent of the total sentenses 
#which has the max score
select_length = int(len(sentence_tokens)*0.4)
select_length
```

```python
summary=nlargest(select_length,sentense_scores,key=sentense_scores.get)
```

```python
#Summary of the input text
summary
```

```python
#Combing the sentenses together...
final_summary=[word.text for word in summary]
```

```python
summary=' '.join(final_summary)
```
